### Q2a) Show the usage of your decision tree on the given dataset. The first 70% of the data should be used for training purposes and the remaining 30% for test purposes. Show the accuracy, per-class precision and recall of the decision tree you implemented on the test dataset.  

*Answer:*  
The decision tree was trained on 70% of the dataset and tested on the remaining 30%.  

- Using *Information Gain* as the splitting criterion:  
  - Accuracy: *0.90*  
  - Class 0 → Precision: *0.9091, Recall: **0.8333*  
  - Class 1 → Precision: *0.8947, Recall: **0.9444*  

- Using *Gini Index* as the splitting criterion:  
  - Accuracy: *0.90*  
  - Class 0 → Precision: *0.9091, Recall: **0.8333*  
  - Class 1 → Precision: *0.8947, Recall: **0.9444*  

This shows that both criteria achieved identical performance, with high accuracy and balanced precision/recall across both classes.

### Q2b) Use 5-fold cross-validation on the dataset. Using nested cross-validation find the optimum depth of the tree.  
A 5-fold nested cross-validation was performed to determine the optimal tree depth.  

- With *Information Gain*:  
  - Depth 2 achieved the best accuracy (*0.92 ± 0.06*)  
  - Accuracy decreased for deeper trees (down to 0.84 at depth 7–8).  
  - *Optimal Depth: 2*  

- With *Gini Index*:  
  - Depth 2 again achieved the best accuracy (*0.92 ± 0.06*)  
  - Accuracy gradually decreased with deeper trees (down to 0.83 at depth 8).  
  - *Optimal Depth: 2*  

Thus, *depth = 2* was consistently optimal across both splitting criteria, showing that shallow trees generalized best while deeper trees tended to overfit.
